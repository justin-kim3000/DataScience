# 키워드, 카테고리, 출력 범위, 정렬기준 설정

```python
search = '해킹'
category = 'book'
display = 20
start = 1
sort = 'sim'
```

```python
baseURL = 'https://openapi.naver.com/v1/search/'
node = f'{category}.json'
param = f'?query={urllib.parse.quote(search)}'
param += f'&display={display}'
param += f'&start={start}'
param += f'&sort={sort}'
URL = baseURL + node + param
```

```python
request = urllib.request.Request(URL)
request.add_header("X-Naver-Client-Id",'jZmv5fKEpC2UFb0IHTwx')
request.add_header("X-Naver-Client-Secret",'T1Wdnw90BT')
response = urllib.request.urlopen(request)
rescode = response.getcode()

if(rescode==200):
    response_body = response.read()
    res_data = response_body.decode('utf-8')
else:
    print("Error Code:" + rescode)
```

```python
# 역슬레시 제거
res_data = re.sub('\\\\', '', res_data)
```

# 파일 생성

```python
file_name = f'{search}_naver_{category}.json'
try:
    with open(file_name,'w', encoding = 'utf-8') as outfile:
        outfile.write(res_data)
        print('json 파일이 저장되었습니다.')
except BaseException as ex:
    print("파일 저장 에러",ex)
```

```python
file_name2 = f'{search}_naver_{category}.csv'
try:
    with open(file_name2,'w', encoding = 'utf-8') as outfile:
        outfile.write(res_data)
        print('csv 파일이 저장되었습니다.')
except BaseException as ex:
    print("파일 저장 에러",ex)
```
