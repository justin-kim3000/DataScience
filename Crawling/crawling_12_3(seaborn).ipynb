```python
# 시각화
from matplotlib import pyplot as plt
# 행렬, 선형대수
from numpy.random import randn
import numpy as np
import seaborn as sns
%config Completer.use_jedi = False

def korean_font():
    plt.rc('font', family='AppleGothic')
    plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot1():
    iris = sns.load_dataset('iris')
    print(iris.info())
    x = iris.petal_length.values
    print(type(x))
    sns.rugplot(x) # x축 위 작은 선분(rug)로 표시
    plt.title('붓꽃 꽃잎 길이')
    plt.show()
    
    sns.kdeplot(x) # kernel density(밀도)
    plt.title('붓꽃 꽃잎 길이 kde')
    plt.show()
    
    sns.distplot(x, kde = True, rug=True)
    plt.title('붓꽃 꽃잎 길이 kde + rug')
    plt.show()
    
```

```python
def plot2():
    titan = sns.load_dataset('titanic')
    print(titan.info())
    print(titan.describe())
    sns.countplot(x='survived',hue='sex',data = titan)
    plt.title('타이타닉호 각 클래스별, 승객 수')
    plt.show()
```

```python
def plot3():
    tips = sns.load_dataset('tips')
    print(tips.info())
    print(tips.describe())
    sns.countplot(x='day',data = tips)
    plt.title('요일별 팁 준 횟수')
    plt.show()
```

```python
def plot4():
    iris = sns.load_dataset('iris')
    print(iris.info())
    print(iris.describe())
    sns.jointplot(x='sepal_length', y='sepal_width',data=iris,kind='kde')
    plt.suptitle('아이리스')
    plt.grid(True)
    plt.show()
```

```python
def plot5():
    iris = sns.load_dataset('iris')
    sns.pairplot(iris)
    plt.suptitle('붓꽃 다차원 차트')
    plt.show()
    
    sns.pairplot(iris, hue='species', markers=['o','s','D'])
    plt.title('품종별 색상 시각화')
    
```
