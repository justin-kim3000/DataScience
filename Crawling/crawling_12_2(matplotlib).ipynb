```python
# 시각화
from matplotlib import pyplot as plt
# 행렬, 선형대수
from numpy.random import randn
import numpy as np
%config Completer.use_jedi = False
```

```python
def korean_font():
    plt.rc('font', family='AppleGothic')
    plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot1():
    plt.title('시각화')
    plt.plot([1,2,3,4,5])
    plt.show()
```

```python
def plot2():
    plt.title('시간별 온도 정보')
    plt.plot([1,2,3,4],[10,20,30,40],'rs--')
    plt.xlabel('시간')
    plt.ylabel('온도')
    plt.show()
```

```python
def plot3():
    x1 = np.linspace(0.0,5.0)
    x2 = np.linspace(0.0, 2.0)
    y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
    y2 = np.cos(2*np.pi*x2)
    plt.subplot(2,1,1)
    plt.plot(x1,y1,'yo-')
    plt.title('subplot 첫번째')
    plt.xlabel('x값')
    plt.ylabel('y값')
    
    plt.subplot(2,1,2)
    plt.plot(x2,y2,'yo-')
    plt.title('subplot 첫번째')
    plt.xlabel('x값')
    plt.ylabel('y값')
    
    plt.tight_layout() # 위, 아래 간격 자동 조절
    plt.show()
    
```

```python
def plot4():
    np.random.seed(0)
    plt.subplot(221)
    plt.plot(np.random.rand(5))
    plt.title('subplot 1')
    
    plt.subplot(222)
    plt.plot(np.random.rand(5))
    plt.title('subplot 2')
    
    plt.subplot(223)
    plt.plot(np.random.rand(5))
    plt.title('subplot 3')
    
    plt.subplot(224)
    plt.plot(np.random.rand(5))
    plt.title('subplot 4')
    
    plt.tight_layout() # 위, 아래 간격 자동 조절
    plt.show()
```

```python
def plot5():
    fig, axes = plt.subplots(2,2)
    np.random.seed(0)
    axes[0,0].set_title('subplot 1번째')
    axes[0,0].plot(np.random.rand(5))
    
    axes[0,1].set_title('subplot 2번째')
    axes[0,1].plot(np.random.rand(5))
    
    axes[1,0].set_title('subplot 3번째')
    axes[1,0].plot(np.random.rand(5))
    
    axes[1,1].set_title('subplot 4번째')
    axes[1,1].plot(np.random.rand(5))
    
    plt.tight_layout() # 위, 아래 간격 자동 조절
    plt.show()
```

```python
def plot6():
    git, ax0 = plt.subplots()
    ax1 = ax0.twinx() # x값을 2개의 subplot 공유
    ax0.set_title('2개의 y축을 한개의 figure에서 사용')
    ax0.plot([10,5,2,9,7], '-r')
    ax0.set_ylabel('y0 값')
    ax0.grid(True)
    
    ax1.plot([100,50,20,90,70], '-r')
    ax1.set_ylabel('y1 값')
    ax1.grid(False)

    plt.tight_layout() # 위, 아래 간격 자동 조절
    plt.show()
        
```
