```python
# 시각화
from matplotlib import pyplot as plt
# 행렬, 선형대수
from numpy.random import randn
import numpy as np
```

```python
def plot1():
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.plot(randn(50).cumsum(),'k--')
    plt.show()
```

```python
def plot2():
    fig = plt.figure()
    ax1 = fig.add_subplot(3,2,1) # subplot(3행 2열 첫번째)
    ax2 = fig.add_subplot(3,2,2)
    ax3 = fig.add_subplot(3,2,3)
    ax4 = fig.add_subplot(3,2,4)
    ax5 = fig.add_subplot(3,2,5)
    ax6 = fig.add_subplot(3,2,6)
    
    ax1.plot(randn(50).cumsum(),'k--') # 검정 점선
    ax2.plot(randn(50).cumsum(),'r')
    ax3.plot(randn(50).cumsum(),'g--') # 녹색 점선
    ax4.plot(randn(50).cumsum(),'k--') # 검정 점선
    ax5.plot(randn(50).cumsum(),'r')
    ax6.plot(randn(50).cumsum(),'g--') # 녹색 점선
    
    plt.show()
```

```python
def plot3():
    fig = plt.figure()
    ax1 = fig.add_subplot(1,1,1) # subplot
    # alpha => 투명도
    ax1.hist(randn(100), bins = 10, color='r', alpha=0.3)
    plt.show()
```

```python
def plot4():
    fig = plt.figure()
    ax1 = fig.add_subplot(1,1,1) # subplot
    ax1.scatter(np.arange(50),np.arange(50)+100*randn(50))
    plt.show()
```

```python
def plot5():
    fig = plt.figure()
    fig.set_size_inches(18.5,10.5)
    
    ax1 = fig.add_subplot(3,2,1) # subplot(3행 2열 첫번째)
    ax2 = fig.add_subplot(3,2,2)
    ax3 = fig.add_subplot(3,2,3)
    ax4 = fig.add_subplot(3,2,4)
    ax5 = fig.add_subplot(3,2,5)
    ax6 = fig.add_subplot(3,2,6)
    
    ax1.plot(randn(30).cumsum(), 'k--')
    ax2.plot(randn(30).cumsum(), 'ko--') #black + marker
    ax3.plot(randn(30).cumsum(), 'yo-', markeredgecolor='r')
    ax4.plot(randn(30).cumsum(), 'ko--', markeredgecolor='y')
    ax5.plot(randn(30).cumsum(), label = 'aa')
    ax5.legend(loc='upper right')
    ax6.plot(randn(30).cumsum(), drawstyle='steps-post',label='계단식 PLOT')
    plt.show()
```
