```python
# 네이버 영화 조회 순위
URL = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
res = requests.get(URL, headers = {'User-Agent':'Mozilla/5.0'})
html = res.text
soup = BeautifulSoup(html, 'html.parser')
rank = 1
for i in soup.select('div[class=tit3]'):
    
    print(str(rank)+". " + i.text.strip())
    rank += 1
```
