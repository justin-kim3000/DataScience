```python
# 시각화
from matplotlib import pyplot as plt
# 행렬, 선형대수
from numpy.random import randn
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris # 머신러닝 lib
%config Completer.use_jedi = False

def korean_font():
    plt.rc('font', family='AppleGothic')
    plt.rcParams['axes.unicode_minus'] = False
```

```python
def plot1():
    iris = load_iris()
    data = iris['data']
    feature = iris['feature_names']
    label = iris['target']
    df = pd.DataFrame(data, columns=feature)
    # add label
    df.loc[:,'label']= label
    
    zero = df[df['label']==0].index
    one = df[df['label']==1].index    
    two = df[df['label']==2].index    
    print(zero)
    
    df.loc[zero,'color']='r'
    df.loc[one,'color']='g'
    df.loc[two,'color']='b'
    df.plot.scatter(x='sepal length (cm)',
                   y='label',c=df['color'])
    plt.show()
    ax = df['sepal length (cm)'].plot.line(color='green')
    ax_2 = df['petal length (cm)'].plot.bar(ax=ax, color='red')
    ax_2.set_xticks(df.index[0::10]) # x축 간격
    ax_2.set_xticklabels(df.index[0::10], rotation=45) # x축 간격 라벨
    ax_2.set_ylabel('legnth(cm)') # y축 라벨
    ax_2.legend()
    plt.plot(data=ax_2)
    plt.show()
    
```

```python
if __name__ == '__main__':
    korean_font()
    plot1()
#     plot2()
#     plot3()
#     plot4()
#     plot5()
```
