```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
#막대 그래프
sns.barplot(x='class', y='fare',data = titanic)
```

```python
# 포인트 플롯
sns.pointplot(x='class',y ='fare',data = titanic)
```

```python
sns.boxplot(x='class', y='age', data=titanic)
```

```python
# 바이올린플롯
sns.violinplot(x='class',y='age',data=titanic)
```

```python
sns.violinplot(x='class',y='age',data=titanic,hue='sex',split=True)
```

```python
# 카운트플롯
sns.countplot(x='class',data=titanic)
```

```python
sns.countplot(y='class',data=titanic)
```

```python
# 히트맵 : 비교데이터가 많을때 사용
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights_pivot = flights.pivot(index='month',
                              columns='year',
                              values='passengers')
flights_pivot
```

```python
sns.heatmap(data=flights_pivot)
```

```python
# 라인플롯 = 두 수치형 데이터 사이의 관계를 나타낼 때 사용. (음영은 95% 신뢰구간 / y는 평균 승객수 나타냄)
sns.lineplot(x='year',y='passengers',data=flights)
```

```python
# 산점도(scatterplot) => 두 데이터의 관계를 점으로 표현하는 그래프
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='time',data=tips)
```

```python
# 회귀선을 포함한 산점도 그래프(regplot) -> 산점도와 선형 회귀선을 동시에 그려줌
# 음영은 95% 신뢰구간
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
# 신뢰구간 99%
sns.regplot(x='total_bill',y='tip',data=tips,ci=99)
```
