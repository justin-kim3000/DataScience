```python
# 푸아송 분포: 단위 시간에 어떤 사건이 몇번 일어난것인지 표현
from math import factorial, exp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pois_dist(n, lamb):
  return (lamb **n) * exp(-lamb) / factorial(n)

def poisson_f():
  x = np.arange(40)
  pd = np.array([pois_dist(n, 10)for n in range(40)])
  plt.ylim(0, 0.15)
  plt.text(33.5, 0.14, 'lamb = 10')
  plt.bar(x, pd, color='blue')
  plt.show()

poisson_f()
```

```python
# 카이제곱 분포
from scipy.stats import chi2
```

```python
dof =2  # 자유도
x = np.linspace(0, 20, 200)
y = chi2(dof).pdf(x)
plt.plot(x,y, 'r--')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title('chi2')
plt.show()
```

```python
#!pip install yfinance
import yfinance as yf
import statsmodels.api as sm
stock_price = yf.download({'005930.KS', '^KS11'},
                          '2022-01-01','2022-10-21')
stock_price.info()
```

```python
stock_price.describe()
```

```python
stock_price.corr()
```

```python
# 종속변수
target = stock_price['Adj Close']['^KS11']
print(target)
x= stock_price['Adj Close']['005930.KS']
```

```python
# 상수값을 더해서 회귀식을 잘 표현
feature_add = sm.add_constant(x, has_constant='add')
model = sm.OLS(target, x)
fittedModel = model.fit()

print('결정계수:', fittedModel.rsquared)
print('p-value:', fittedModel.pvalues)
print('잔차의 제곱합:', fittedModel.mse_resid)
target_pred = fittedModel.predict()

plt.scatter(x, target, marker ='+', label = 'Actual value')
plt.plot(x, target_pred, color='g',label='Predicted value')
plt.hlines(target.mean(), x.min(), x.max(), color='r',label='mean')
plt.xlabel('SAM')
plt.ylabel('KOSPI')

plt.legend()
```
