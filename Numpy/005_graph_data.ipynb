```python
# 나이 추가하는 법
df['나이'] = 2019-df['출생년도']
```

```python
df = df.astype({'나이':'int'})
df
```

```python
df = df.assign(나이구간=np.where(df['나이']<30,'청년',
                                np.where(df['나이']<=45,'장년',
                                np.where(df['나이']<=64,'중년','노인'))))
```

```python
# df = df.rename(columns={'testage':'나이구간'})
# df= df.drop('나이구간',axis=1)
```

```python
sns.countplot(data=df, x='나이구간')
```

```python
sns.histplot(data=df, x='나이')
```

```python
age_salary2 = df.dropna(subset=['급여']).groupby(['나이구간','성별'],as_index=False).agg(mean_income=('급여','mean'))
age_salary2
```

```python
sns.barplot(data=age_salary2, x='나이구간',y='mean_income',hue='성별',order=['청년','장년','중년','노인'])
```

```python
age_salary = df.dropna(subset=['급여']).groupby('나이',as_index=False).agg(mean_income=('급여','mean'))
age_salary
```

```python
# 스텝 1: m행 n열 Figure 준비
# 폰트 크기 설정
mpl.rc('font', size=11)
# 각 축의 제목 크기 설정
mpl.rc('axes', titlesize=15)
# 3행 2열 Fighure 생성
figure, axes = plt.subplots(2,1)
# 그래프 사이에 여백 확보
plt.tight_layout()
# 전체 Figure 크기를 10x9 인치로 설정
figure.set_size_inches(15, 10)
sns.barplot(data=age_salary,x='나이',y='mean_income',ax=axes[0])
sns.lineplot(data=age_salary,x='나이',y='mean_income',ax=axes[1])
# 3-2 : 1행에 위치한 서브플롯들의 x축 라벨 90도 회전
# (회전 하지않을 경우 숫자가 겹침)
axes[0].tick_params(axis='x', labelrotation=40)
```
