```python
sns.barplot(data=marri_about, x='결혼여부',y='cnt')
```

```python
# rel_div = df.query('결혼여부 !=""').groupby('종교', as_index=False)['결혼여부'].value_counts(normalize=True)
rel_div = df.query('결혼여부 != ""').groupby('종교')['결혼여부'].value_counts(normalize=True).reset_index(name='counts')

rel_div
```

```python
sns.barplot(data=rel_div, x='종교',y='counts',hue='결혼여부')
```

```python
rel_div = rel_div.query('결혼여부=="이혼"').assign(proportion=rel_div['counts']*100).round(1)
rel_div
# sns.barplot(data=rel_div, x='종교',y='counts')
```

```python
# 연령대 종교 유무에 따른 이혼율
age_div = df.query('결혼여부 !="기타"').groupby('나이구간')['결혼여부'].value_counts(normalize=True).reset_index(name='proportion')
age_div.sort_values('proportion',ascending=False)
```

```python
age_rel_div = df.query('결혼여부 != "기타" & 나이구간 != "청년"').groupby(['나이구간','종교'])['결혼여부'].value_counts().reset_index(name='proportion')
age_rel_div
```

```python
sns.barplot(data=age_rel_div,x='나이구간',y='proportion',hue='종교')
```

### 지역별 노인 인구 비율

```python
df['지역코드']
```

```python

df_city = pd.DataFrame({'지역코드':[1,2,3,4,5,6,7],
                        '지역명':['서울','수도권(인천/경기)','부산/경남/울산','대구/경북','대전/충남','강원/충북','광주/전남/전북/제주도']})
df_city
```

```python
# 지역 코드에 맞춰서 붙이기
df = df.merge(df_city, how='left', on='지역코드')
```

```python
df[['지역코드','지역명']]
```

```python
df_loc = df.groupby('지역명')['나이구간'].value_counts(normalize=True).reset_index(name='proportion')
sns.barplot(data=df_loc, y='지역명',x='proportion',hue='나이구간')
```

```python
pivot_df = df_loc[['지역명','나이구간','proportion']].pivot(index='지역명',columns='나이구간',values='proportion')
pivot_df
```

```python
pivot_df.plot.barh(stacked=True)
```

```python
from numpy.core.shape_base import stack
sort_df = pivot_df.sort_values(by='노인', ascending = True)[['청년','장년','중년','노인']]
sort_df.plot.barh(stacked=True)
```
